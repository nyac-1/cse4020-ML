```python
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np

dataset, classes = make_blobs(n_samples=300, n_features=2, centers=5, cluster_std=0.4, random_state=2)
df = pd.DataFrame(dataset, columns=['x1', 'x2'])
X = df.values
x1 = df['x1']
x2 = df['x2']
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5)
model = ac.fit(X)
```

```python
print(model.n_clusters)
print(model.n_leaves_)
print(model.n_features_in_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x1,x2, hue = model.labels_)
```
